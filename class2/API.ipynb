# Implementacion de API

```python
# Importando los datos de prueba
import pandas as pd

# Datos prueba
X_test = pd.read_csv("X_test.csv", index_col=0)
y_test = pd.read_csv("y_test.csv", index_col=0)

# Unir ambos en un df
data_test = pd.concat([X_test, y_test], axis=1)
```

```python
data_dict = data_test.to_dict(orient='list')

data_test
```

```python
import json

# Formateado para la API:
data_json = json.dumps({"data": [data_dict]})

suri = open("uri.json", "r")
scoring_uri = json.load(suri)["URI"][0]
suri.close()
```

```python
# La API per-sÃ©:

headers = {"Content-Type": "application/json"}
import requests

response = requests.post(scoring_uri, data=data_json, headers=headers)

if response.status_code == 200:
  result = json.loads(response.json())
  print(result)
  data_test["bankrupt"] = result
  display(data_test)
else:
  print(f"Error: {response.text}")
```
