```python
# Librerías
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import math
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("C:\\Users\\sarah\\Downloads\\bankruptcy\\data.csv")
```

```python
df.head()
```

```python
# Limipar duplicados
df.drop_duplicates

# Limpiar nombres de columnas 
df.columns = df.columns.str.strip().str.replace(" ", "")

# Eliminar valores nulos 
df.dropna(inplace=True)

# Calcular la matriz de correlación
corr_matrix = df.corr()

# Obtener la correlación de cada variable con 'Bankrupt?'
correlation_with_target = corr_matrix["Bankrupt?"].abs().sort_values(ascending=False)

# Seleccionar las 10 variables más correlacionadas (excluyendo 'Bankrupt?' en sí)
top_features = correlation_with_target[1:11]  
print(top_features)

top_features_array = top_features.index.to_numpy()
print(f"Array de columnas: ",top_features_array)

```
